```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
dataset=pd.read_csv("50_Startups.csv")
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
dataset
```

```python
print(X,"\n")
print(y)
```

```python
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3])], remainder='passthrough')
X=np.array(ct.fit_transform(X))
X
```

```python
#X=X[:,1:]#here not ommitted but you can ommit one dummy var
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
np.set_printoptions(precision=2)
```

```python
acc=r2_score(y_test,y_pred)
print("Accuracy=",acc*100,"%")
```

```python
#axis = 0 means horizontal concatenation axis = 1 means vertical concatenation
#reshape is used to represent the y_pred and y_test vertically

print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape((len(y_test),1))),axis=1))
```

Making a single prediction (for example the profit of a startup with R&D Spend = 160000, Administration Spend = 130000, Marketing Spend = 300000 and State = 'California')

```python
print(regressor.predict([[1,0,0,160000,130000,300000]]))
```

```python
print("slope(m)=",regressor.coef_)
print("intercept(c)=",regressor.intercept_)
```
